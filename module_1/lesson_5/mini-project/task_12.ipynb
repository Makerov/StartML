|||mean_price|mean_rating|mean_reviews|

Колонки должны иметь соответствующие названия, значения таблицы округлены до указанного числа знаков после точки. В качестве ответа загрузите полученную таблицу в формате csv (сепаратор запятая - `,`).

**NB!** Понадобится погуглить информацию по запросу "Сводная таблица pandas"

**NB!** Таблицу нужно делать по данным, где удалены дубликаты

```python
import pandas as pd

playstore = pd.read_csv('playstore.csv', index_col=0)
playstore = playstore  \
    .drop_duplicates(subset='App')  \
    .reset_index(drop=True)  \
    .rename(columns=lambda x: x.lower().replace(' ', '_'))
playstore['price'] = playstore['price'].str.replace('$', '').astype('float')
playstore.info()
playstore.head()
```

```python
playstore.groupby(['category', 'type'])  \
    .agg({
        'price': 'mean',
        'rating': 'mean',
        'reviews': 'mean'
    })  \
    .rename(columns=lambda x: f'mean_{x}')  \
    .round(2)
    
```

```python
result = pd.pivot_table(
    playstore,
    index=['category', 'type'],
    values=['price', 'rating', 'reviews'],
    aggfunc='mean'
)  \
    .rename(columns=lambda x: f'mean_{x}')
result
```

```python
result['mean_price'] = result['mean_price'].round(2)
result['mean_rating'] = result['mean_rating'].round(1)
result['mean_reviews'] = result['mean_reviews'].round(2)
```

```python
result.to_csv('task_12_result.csv')
```
