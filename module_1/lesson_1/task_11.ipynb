# \> Коллекции: записная книжка

Теперь, когда мы определились со структурой данных, мы можем создать свою записную книжку.
Ее содержание должно быть следующим: 

|Приоритет|Задача|
|-|-|
|1|Полить цветы|
|0|Покормить кота|
|1|Забрать посылку|
|2|Почитать книгу по программированию|
|3|Ответить на письмо двоюродной тети|

**Задача:** Напишите код, создающий словарь, и сохраните его в переменную `tasks`.

NB! Учтите, что задачи одинакового приоритета должны быть в том же порядке, что и в таблице!

NB! Ключи в этом словаре должны иметь целочисленный тип.

Это может быть немного трудно, поэтому для решения задачи можно повторить:

- Как создавать список
- Как создавать ключ в словаре

Выполняя код в юпитер ноутбуке, вы можете выводить промежуточный результат через `print(tasks)` — это удобный способ следить, на правильном ли вы пути.

```python
tasks = {
    0: ['Покормить кота'],
    1: ['Полить цветы', 'Забрать посылку'],
    2: ['Почитать книгу по программированию'],
    3: ['Ответить на письмо двоюродной тети']
}
tasks
```
